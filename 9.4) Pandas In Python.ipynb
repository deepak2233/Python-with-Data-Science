### Pandas Data Structure
Pandas can read three types of Data Structure.
- <b>Series:</b> A Series, by contrast, is a sequence of data values. If a DataFrame is a table, a Series is a list. And in fact you can create one with nothing more than a list:

- <b>DataFrame:</b> A DataFrame is a table. It contains an array of individual entries, each of which has a certain value. Each entry corresponds to a row (or record) and a column.

- <b>Panel:</b> It is a 3 dimensional data type 

<b>Creating series from list</b>

```python
list_1 = [1, 2, -3, 4.5, 'indian']
print(list_1)
```

<b> Creating series using the above list</b>


```python
series1 = pd.Series(list_1)
print(series1)
```

```python
#Series
sr = pd.Series([1, 2, 3, 4, 5])
print(sr)
# Here it is showing 0 1 2 3 4 is index and 1 2 -3 4.5 Indian are data values
```

```python
# for type
print(type(sr))
```

<b> Creating Series using a different method </b>

```python
# We got an error because we passed 3 indexes for 5 data values So both are eaual
series2 = pd.Series([1,2,3,4,5], index = ['a', 'b', 'c', 'd', 'e'])
print(series2)
```

<b>Changing data type of series (Convert int into a float)</b>

```python
series2 = pd.Series([1,2,3,4,5], index = ['a', 'b', 'c', 'd', 'e'], dtype = float)
print(series2)
```

<b> Creating series from python dictionary</b>

```python
s4_dict = pd.Series({'a':1, 'b':2, 'c':3})
print(s4_dict)
```

<b> Adding two serieses</b>

```python
s5 = pd.Series([1,2,3,4,5])
s6 = pd.Series([1,2,3,4,5])
 
a = s5 + s6
print(a)
```
